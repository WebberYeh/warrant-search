```python
import numpy as np

df = pd.DataFrame(np.r_[tables[0].values[0:,0:2], tables[0].values[0:,2:4], tables[0].values[0:,4:6]],columns=["標的代碼","公司名稱"])
df2 = pd.DataFrame(np.r_[tables[1].values[0:,0:2], tables[1].values[0:,2:4], tables[1].values[0:,4:6]],columns=["標的代碼","公司名稱"])

df3 = pd.DataFrame(np.r_[tables2[0].values[0:,0:2], tables2[0].values[0:,2:4], tables2[0].values[0:,4:6]],columns=["標的代碼","公司名稱"])
df4 = pd.DataFrame(np.r_[tables2[1].values[0:,0:2], tables2[1].values[0:,2:4], tables2[1].values[0:,4:6]],columns=["標的代碼","公司名稱"])
```

```python
print(df)
print(df2)
```

```python
dfs=df.dropna()
dfs2=df2.dropna()
dfs3=df3.dropna()
dfs4=df4.dropna()
print(dfs)
print(dfs2)
```

```python
print(dfs.dtypes)
```

```python
dfs["標的代碼"] = dfs["標的代碼"].astype(int).astype(str)
dfs2["標的代碼"] = dfs2["標的代碼"].astype(int).astype(str)
dfs3["標的代碼"] = dfs3["標的代碼"].astype(int).astype(str)
dfs4["標的代碼"] = dfs4["標的代碼"].astype(int).astype(str)
```

```python
print(dfs["標的代碼"][92])
print(dfs["公司名稱"][92])
print(dfs["標的代碼"][134])
print(dfs["公司名稱"][134])

print(dfs3["標的代碼"][138])
print(dfs3["公司名稱"][138])
print(dfs3["標的代碼"][169])
print(dfs3["公司名稱"][169])
```

```python
dfs["公司名稱"][92] = "恒耀"
dfs["公司名稱"][134] = "宏碁資"
dfs3["公司名稱"][138] = "啟碁"
dfs3["公司名稱"][169] = "宏碁"
```

```python
print(dfs["公司名稱"][92])
print(dfs["公司名稱"][134])
print(dfs3["公司名稱"][138])
print(dfs3["公司名稱"][169])
```

```python
dfall = pd.concat([dfs,dfs2],axis=0)
dfall2 = pd.concat([dfs3,dfs4],axis=0)
```

```python
dfall = dfall.sort_values(by=['標的代碼'])
dfall.index = range(len(dfall))
dfall
```

```python
dfall2 = dfall2.sort_values(by=['標的代碼'])
dfall2.index = range(len(dfall2))
dfall2
```
